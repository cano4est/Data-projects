```python
df_categories = pd.read_sql('select * from categories', engine)
df_customers = pd.read_sql('select * from customers', engine)
df_products = pd.read_sql('select * from products', engine)
df_suppliers = pd.read_sql('select * from suppliers', engine)
df_orders = pd.read_sql('select * from orders', engine)
df_order_details = pd.read_sql('select * from order_details', engine)
```

# Transformación:
- Eliminamos duplicados de las dimensiones
- Construimos la tabla de hechos
- Construimos las dimensiones necesarias

Dimensión Categorías:
- Parte de df_categories
- Eliminar la columna picture

```python
dim_categories = df_categories.drop('picture', axis = 1)
dim_categories
```

Dimensión Clientes:
- Mantener de df_customers id, nombre, contacto, ciudad y region

```python
dim_customers = df_customers[['customer_id', 'company_name', 'contact_name', 'city', 'region']]
dim_customers.head()
```

Dimensión Productos:
- Mantener id, nombre, precio y descontinuado

```python
dim_products = df_products[['product_id', 'product_name', 'unit_price', 'discontinued']]
dim_products.head()
```

Dimensión Proveedor:
- Mantener id, nombre, contacto, ciudad y región

```python
dim_suppliers = df_suppliers[['supplier_id', 'company_name', 'contact_name', 'city', 'region']]
dim_suppliers.head()
```

Dimensión Temporal:
- Fecha
- Mes
- Día
- Año
- Nombre mes
- Nombre día
- Trimestre

```python
dim_time = df_orders[['order_date']]
dim_time['order_date'] = pd.to_datetime(dim_time['order_date'], format = '%Y/%m/%d')
dim_time['month'] = [i.month for i in dim_time['order_date']]
dim_time['day'] = [i.day for i in dim_time['order_date']]
dim_time['year'] = [i.year for i in dim_time['order_date']]
dim_time['month_name'] = [i.strftime('%B') for i in dim_time['order_date']]
dim_time['day_name'] = [i.strftime('%A') for i in dim_time['order_date']]
dim_time['quarter'] = [i.quarter for i in dim_time['order_date']]
dim_time.rename(columns = {'order_date':'date'}, inplace = True)
dim_time = dim_time.drop_duplicates(ignore_index = True)
dim_time.head()
```

Hechos Órdenes:
- Número de orden (order)
- Costo de movilización (freight)
- Precio unitario (unit_price)
- Cantidad de producto (quantity)
- Descuento (discount)
- Monto de venta (amount)
- ID de las dimensiones


```python
fact_orders = df_orders[['order_id', 'customer_id', 'order_date', 'freight']]
fact_orders = fact_orders.merge(df_order_details, how = 'inner', left_on = 'order_id', right_on = 'order_id')
fact_orders['amount'] = fact_orders['unit_price'] * fact_orders['quantity']
fact_orders = fact_orders.merge(df_products, how = 'inner', left_on = 'product_id', right_on = 'product_id')
fact_orders = fact_orders[['order_id', 'freight', 'unit_price_x', 'quantity',
                           'discount', 'amount', 'customer_id', 'category_id', 'order_date',
                           'supplier_id', 'product_id']]
fact_orders.rename(columns = {'unit_price_x':'unit_price'}, inplace = True)
fact_orders.rename(columns = {'order_id':'orders'}, inplace = True)
fact_orders['order_id'] = fact_orders.index + 1
fact_orders
```

# Carga:
- Identificar el objetivo analítico de datos
- Ejecutar el proceso de cargue

Crear los archivos planos de las tablas de hechos y dimensiones

```python
dim_categories.to_csv('dim_categories.csv', sep = '|', index = False)
dim_products.to_csv('dim_products.csv', sep = '|', index = False)
dim_customers.to_csv('dim_customers.csv', sep = '|', index = False)
dim_suppliers.to_csv('dim_suppliers.csv', sep = '|', index = False)
dim_time.to_csv('dim_time.csv', sep = '|', index = False)
fact_orders.to_csv('fact_orders.csv', sep = '|', index = False)
```

Crear las tablas en NorthWindDW

```python
connection = psycopg2.connect(host = 'localhost',
                              database = 'NorthWindDW',
                              user = 'postgres',
                              password = 'admin')
cursor = connection.cursor()
cursor.execute("""CREATE TABLE IF NOT EXISTS dim_categories(category_id integer, category_name text, description text)""")
cursor.execute("""CREATE TABLE IF NOT EXISTS dim_customers(customer_id text, company_name text, contact_name text, city text, region text)""")
cursor.execute("""CREATE TABLE IF NOT EXISTS dim_suppliers(supplier_id integer, company_name text, contact_name text, city text, region text)""")
cursor.execute("""CREATE TABLE IF NOT EXISTS dim_time(date date, month integer, day integer, year integer, month_name text, day_name text, quarter integer)""")
cursor.execute("""CREATE TABLE IF NOT EXISTS dim_products(product_id integer, product_name text, unit_price float, discontinued integer)""")
cursor.execute("""CREATE TABLE IF NOT EXISTS fact_orders(orders integer, freight float, unit_price float, quantity integer, discount float, amount float, customer_id text, category_id integer, order_date date, supplier_id integer, product_id integer, order_id integer)""")
cursor.execute("""ALTER TABLE ONLY dim_categories ADD CONSTRAINT pk_category_id PRIMARY KEY (category_id)""")
cursor.execute("""ALTER TABLE ONLY dim_customers ADD CONSTRAINT pk_customer_id PRIMARY KEY (customer_id)""")
cursor.execute("""ALTER TABLE ONLY dim_suppliers ADD CONSTRAINT pk_supplier_id PRIMARY KEY (supplier_id)""")
cursor.execute("""ALTER TABLE ONLY dim_time ADD CONSTRAINT pk_date PRIMARY KEY (date)""")
cursor.execute("""ALTER TABLE ONLY dim_products ADD CONSTRAINT pk_product_id PRIMARY KEY (product_id)""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT pk_order_id PRIMARY KEY (order_id)""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT fk_category_id FOREIGN KEY (category_id) REFERENCES dim_categories""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT fk_customer_id FOREIGN KEY (customer_id) REFERENCES dim_customers""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT fk_supplier_id FOREIGN KEY (supplier_id) REFERENCES dim_suppliers""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT fk_date FOREIGN KEY (order_date) REFERENCES dim_time""")
cursor.execute("""ALTER TABLE ONLY fact_orders ADD CONSTRAINT fk_product_id FOREIGN KEY (product_id) REFERENCES dim_products""")
```
